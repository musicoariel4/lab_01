## Tarea 1: Palíndromo

Determina si un número entero $n$ es un palíndromo, es decir, si se lee igual de izquierda a derecha que de derecha a izquierda. Devuelve `True` si lo es; de lo contrario, devuelve `False`.

**Ejemplo de ejecución**

```python
>>> palindromo(n=3456543)
True

>>> palindromo(n=543)
False
```

**Reto**: usa una sola línea de código.

````python
def palindromo(n): # n llega como entero

    # Convertir el número a cadena
    num_str = str(n)
    # Comparar con su reverso
    if num_str == num_str[::-1]:
        return True
    else:
        return False
````

````python
print (palindromo(12321))
print (palindromo(12322))
````

## Tarea 2: Cambio

En esta tarea, tu función recibirá una cantidad entre 1 y 99 centavos y deberá devolver una lista con tres números [$n_1$, $n_{10}$, $n_{25}$], que especifican cuántas monedas de 1, 10 y 25 centavos se requieren para obtener la cantidad dada. Solo disponemos de monedas de 1, 10 y 25 centavos. No hay otros tipos de monedas.

Si la cantidad dada es menor que 1 o mayor que 99, debes devolver `None`.

**Ejemplo de ejecución**

```python
>>> change_money(n=55)
[5,0,2]

>>> change_money(n=200)
None

>>> change_money(n=47)
[2,2,1]
```

**Sugerencia**: usa `\\` para la división entera


Comprueba tu código manualmente


````python

def change_money(n):
    r = [0, 0, 0]  # [monedas de 1, 10, 25]

    # Validar rango
    if n < 1 or n > 99:
        return None

    # Monedas de 25
    if n >= 25:
        r[2] = n // 25
        n = n % 25

    # Monedas de 10
    if n >= 10:
        r[1] = n // 10
        n = n % 10

    # Monedas de 1
    if n >= 1:
        r[0] = n

    return r
````

````python
for i in [1, 9, 10, 24, 25, 37, 50, 99]:
    print(i, "->", change_money(i))
````

## Tarea 3: Fibonacci

Complete la siguiente función para que, dado un entero $n$, calcule el término $n^{-ésimo}$ de la serie de Fibonacci:

$$ f_n = f_{n-1}+f_{n-2}$$

Donde $$f_0=0 \:\:\text{ y }\:\:f_1=1 $$ de modo que: $$f_2=1+0=1$$ $$f_3=1+1=2$$ $$f_4=2+1=3$$ y así sucesivamente

### Ejemplo de ejecución
```python

>>> fibonacci(10)
55

>>> fibonacci(36)
14930352
```

````python
def fibonacci(n):
    f_1 = 1
    f_2 = 1
    suma = 0

    if n <= 0:
        return 0

    if n == 1:
        return 1

    # tu código aquí
    suma = f_1 + f_2
    for i in range(3, n + 1):
        f_1, f_2 = f_2, f_1 + f_2
        suma += f_2

    return suma
````

## Tarea 4: Número perfecto

Complete la siguiente función para que acepte un entero $n$ y devuelva `Verdadero` si $n$ es un número perfecto y `Falso` en caso contrario.

Un número perfecto es un entero positivo que es igual a la suma de sus divisores positivos, sin incluir el propio número. Por ejemplo, $6 = 3 + 2 + 1$.

**Ejemplo de ejecución**
```python
>>> perfecto(6)
True

>>> perfecto(51)
False

>>> perfecto(496)
True
```




````python
def perfecto(n):
    suma = 0

    # tu código aquí
    for i in range(1, n):
        if n % i == 0:
            suma += i

    if suma == n:
        return True
    else:
        return False
````
